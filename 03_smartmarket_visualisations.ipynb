```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("output/smartmarket_leads_scope.csv", parse_dates=["date"])
camp = pd.read_csv("output/smartmarket_campaigns_kpi.csv")

plt.rcParams["figure.dpi"] = 120
```

```python
# Quels canaux génèrent le plus de leads ?
vol = df["channel"].value_counts().sort_values(ascending=False)

plt.figure(figsize=(7,4))
vol.plot(kind="bar")
plt.title("Volume de leads par canal (septembre 2025)")
plt.xlabel("Canal")
plt.ylabel("Nombre de leads")
plt.xticks(rotation=25, ha="right")
plt.tight_layout()
plt.savefig("graphs/viz_01_leads_par_canal.png")
plt.show()
```

```python
# Quels canaux génèrent les leads les plus qualifiés ?
share = pd.crosstab(df["channel"], df["status"], normalize="index").sort_index()

plt.figure(figsize=(7,4))
share.plot(kind="bar", stacked=True)
plt.title("Qualité des leads par canal (répartition MQL/SQL/Client)")
plt.xlabel("Canal")
plt.ylabel("Proportion")
plt.xticks(rotation=25, ha="right")
plt.legend(title="Statut", bbox_to_anchor=(1.02, 1), loc="upper left")
plt.tight_layout()
plt.savefig("graphs/viz_02_qualite_par_canal.png")
plt.show()
```

```python
# Sur quels appareils arrivent les leads ?
dev = df["device"].value_counts()

plt.figure(figsize=(6,4))
dev.plot(kind="bar")
plt.title("Répartition des leads par device")
plt.xlabel("Device")
plt.ylabel("Nombre de leads")
plt.xticks(rotation=0)
plt.tight_layout()
plt.savefig("graphs/viz_03_leads_par_device.png")
plt.show()
```

```python
# Quelles régions rapportent le plus de leads ?
reg = df["region"].value_counts().head(10)

plt.figure(figsize=(7,4))
reg.sort_values().plot(kind="barh")
plt.title("Top régions par volume de leads")
plt.xlabel("Nombre de leads")
plt.ylabel("Région")
plt.tight_layout()
plt.savefig("graphs/viz_04_top_regions.png")
plt.show()
```
